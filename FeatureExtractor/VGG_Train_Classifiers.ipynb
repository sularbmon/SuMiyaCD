```python
import pickle
x_train = pickle.load(open('../June_Weights_VGG/GENERAL_X_TRAIN.pkl','rb'))
y_train = pickle.load(open('../June_Weights_VGG/GENERAL_Y_TRAIN.pkl','rb'))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train, y_train)
print('done')
```

```python
knn_filename ='../June_Weights_VGG/VGG_KNN.pkl'
pickle.dump(knn,open(knn_filename,'wb'))

print("done")
```

```python
#RANDOM FOREST
from sklearn.ensemble import RandomForestClassifier
RF_model = RandomForestClassifier(n_estimators = 300, random_state = 72) #75-68%  #100 86%  #72-93%
print("defined rf")
# Train the model on training data
RF_model.fit(x_train, y_train) #For sklearn no one hot encoding
print("RF_model.fit is done")
```

```python
rf_filename ='../June_Weights_VGG/VGG_RF.pkl'
pickle.dump(RF_model,open(rf_filename,'wb'))
print('done')
```

```python
#SVM

from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
from sklearn.calibration import CalibratedClassifierCV

clf = LinearSVC(random_state=21, tol=0.0005,max_iter=1500) #rs=10 maxiter=500  tol=0.01  
clf.fit(x_train, y_train)
print("done")
```

```python
svm_filename ='../June_Weights_VGG/VGG_SVM.pkl'
pickle.dump(clf,open(svm_filename,'wb'))
```

```python
import lightgbm as lgb

lgb_train = lgb.Dataset(x_train, y_train)

# Set LightGBM parameters
params = {
    'objective': 'multiclass',
    'num_class': 147,  # Replace 'num_classes' with the actual number of classes in your dataset
    'metric': 'multi_logloss',
}
model = lgb.train(params, lgb_train, num_boost_round=100)
print('done')
```

```python
model_name ='../June_Weights_VGG/VGG_LIGHTGBM.pkl'
pickle.dump(model,open(model_name,'wb'))

print("done")
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

```python
gnb_filename ='../June_Weights_VGG/VGG_GNB.pkl'
pickle.dump(gnb,open(gnb_filename,'wb'))
print('done')
```
